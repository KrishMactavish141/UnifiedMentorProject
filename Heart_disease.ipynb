#### Importing libraries and dataset and explore rows

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heart=pd.read_csv("D:\\Career\\Unified_Mentor_Intenship\\Heart Disease data\\Heart Disease data.csv")
heart.head(5)
```

#### Checking The Shape of the Dataset

```python
print('The number of rows in the dataset ',heart.shape[0])
print('The number of columns in the dataset ',heart.shape[1])
```

#### The detailing of the dataset

```python
heart.info()
```

###### This indicates there is no null values present in the dataset

#### Checking duplicates in the data

```python
heart.duplicated().sum()
```

#### Get overall Statistics of the dataset

```python
heart.describe(include='all')
```

#### Draw Correlation Matrix

```python
cor=heart.corr()
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```

#### Heart Disease count (Presence or no Presence)?

```python
heart['target'].value_counts()
```

###### This indicates heart disease presence is higher than individuals with no heart disease

#### Which sex has the most heart disease (Male or Female)?

```python
sns.countplot(x='sex',data=heart,hue='target')
plt.legend(labels=['No-Disease','Disease'])
```

###### It is clearly indicates that male (sex value 1) have higer presence of heart disease comapre to female (sex value 0)

#### Check Age Distribution in the Dataset

```python
sns.histplot(heart['age'],bins=20)
plt.show()
```

#### Check Chest Pain Type
<b>Chest pain type (4 values)</b>
###### 0:typical angina
###### 1:atypical angina
###### 2:non-anginal pain
###### 3:asymptomatic


```python
plt.figure(figsize=(8,5))
sns.countplot(x='cp',data=heart,hue='cp')
plt.legend(labels=['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.show()
```

#### Chest Pain Distribution as per Target Variable 

```python
sns.countplot(x='cp',data=heart,hue='target')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

#### Fasting Blood Sugar Distribution According to the Target Variable

```python
sns.countplot(x='fbs',data=heart,hue='target')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

#### Check Resting Blood Pressure Distribution

```python
sns.histplot(heart['trestbps'])
```

#### Comparing Resting Blood Pressure with Sex

```python
f=sns.FacetGrid(heart,hue='sex',aspect=4)
f.map(sns.kdeplot,'trestbps',fill=True)
plt.legend(labels=['Male','Female'])
```
